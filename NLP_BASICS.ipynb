```python
#import libraries

import re
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
import spacy
from pprint import pprint

#load spaCy English model
nlp = spacy.load("en_core_web_sm")
```

```python
#input text

text= "apple is LOOKING at Buying a U.K. startup for $1 billion in 2025"
print("original text: \n",text)

#preprocessing

def preprocess(text):
    text = text.lower() #lowercase
    text = re.sub(r'[^a-zA-Z\s]','',text) #remove special charecters
    return text

clean_text = preprocess(text)
print("preprocessed text: \n",clean_text)

```

```python
#tokenization
tokens=word_tokenize(clean_text)
print("tokens:\n",tokens)

#stopwords
stopwd=set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word not in stopwd]
print("filtered tokens:\n",filtered_tokens)

#stemming
stemmer=PorterStemmer()
stemmed_tokens=[stemmer.stem(word) for word in filtered_tokens]
print("stemmed tokens:\n",stemmed_tokens)

#lemmatization
lemmatizer=WordNetLemmatizer()
lemmatized_tokens=[lemmatizer.lemmatize(word,pos="v") for word in filtered_tokens]
print("lemmatized tokens:\n",lemmatized_tokens)

#pos tagging
print("pos yags:\n")
pprint(nltk.pos_tag(tokens))

#named entity recognition
print("named entity recognition: \n")
doc=nlp(text)
for ent in doc.ents:
    print(ent.text,ent.label)
```
